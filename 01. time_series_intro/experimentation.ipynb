```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, LSTM, GRU, Dense, Dropout
from tensorflow.keras.optimizers import Adam, SGD
from functions import load_and_preprocess_data, plot_results, calculate_rmse

# Load and preprocess data
X_train, X_test, y_train, y_test, scaler = load_and_preprocess_data('airpassengers.csv')

# Define model configurations
models = {
    "SimpleRNN": SimpleRNN,
    "LSTM": LSTM,
    "GRU": GRU
}

results = {}

for model_name, layer in models.items():
    # Build the model
    model = Sequential([
        layer(50, activation='relu', input_shape=(X_train.shape[1], 1)),
        Dense(1)
    ])

    model.compile(optimizer=Adam(learning_rate=0.01), loss='mse')

    # Train the model
    history = model.fit(X_train, y_train, epochs=100, batch_size=16, validation_split=0.2, verbose=1)

    # Make predictions
    train_predict = model.predict(X_train)
    test_predict = model.predict(X_test)

    # Inverse transform predictions
    train_predict = scaler.inverse_transform(train_predict)
    y_train_inv = scaler.inverse_transform(y_train.reshape(-1, 1))
    test_predict = scaler.inverse_transform(test_predict)
    y_test_inv = scaler.inverse_transform(y_test.reshape(-1, 1))

    # Calculate RMSE
    train_rmse, test_rmse = calculate_rmse(train_predict, y_train_inv, test_predict, y_test_inv)

    # Store results
    results[model_name] = {
        "train_rmse": train_rmse,
        "test_rmse": test_rmse,
        "history": history
    }

    # Plot results
    plot_results(history, y_train_inv, train_predict, y_test_inv, test_predict)

for model_name, result in results.items():
    total_params = result['history'].model.count_params()
    print(f"{model_name} - Train RMSE: {result['train_rmse']:.3f}, Test RMSE: {result['test_rmse']:.3f}, Parameters: {total_params}")
```

### Reminder: 

- SimpleRNN: Good for simple, short-term dependencies.
- LSTM: Best for capturing long-term dependencies and complex patterns, thanks to its gating mechanisms.
- GRU: A more efficient alternative to LSTM, often performing similarly with fewer parameters.

```python
def evaluate_model(model, X_train, X_test, y_train, y_test, scaler):
    # Make predictions
    train_predict = model.predict(X_train)
    test_predict = model.predict(X_test)

    # Inverse transform predictions
    train_predict = scaler.inverse_transform(train_predict)
    y_train_inv = scaler.inverse_transform(y_train.reshape(-1, 1))
    test_predict = scaler.inverse_transform(test_predict)
    y_test_inv = scaler.inverse_transform(y_test.reshape(-1, 1))

    # Calculate RMSE
    train_rmse, test_rmse = calculate_rmse(train_predict, y_train_inv, test_predict, y_test_inv)
    
    return train_rmse, test_rmse, train_predict, test_predict, y_train_inv, y_test_inv

# Example usage:
train_rmse, test_rmse, train_predict, test_predict, y_train_inv, y_test_inv = evaluate_model(
    model, X_train, X_test, y_train, y_test, scaler
)
```
